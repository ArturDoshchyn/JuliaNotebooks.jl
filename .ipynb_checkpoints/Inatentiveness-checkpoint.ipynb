```julia
using Parameters, Plots

function V(d, params)
    @unpack γ, ρ, σ, β, κ = params
    if ρ ≈ 1.0
        return (γ * σ^2 * d / (1 - β) - κ) * β^d / (1 - β^d)
    else 
        return (- (γ * σ^2 / (1 - ρ^2)) * ((1 - β^d) / (1 - β) - (1 - (β * ρ^2)^d) / (1 - β * ρ^2)) - β^d * κ) / (1 - β^d)
    end
end
```

```julia
#Parameters
params = (
    γ = 1.0,
    ρ = 0.99,
    σ = 1.0,
    β = 0.96,
    κ = 10
    )

plot(1:25, x -> V(x, params))
```

```julia
V(2, params)
```
